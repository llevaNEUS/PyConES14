#Calcular el valor medio a través de un valor central y una ventana definida (modificado 2)

*Ejemplo con el cálculo del valor medio de reflectancia en un rango definido de longitud de onda*

```python
%matplotlib inline
```

```python
from numpy import arange
import random
import matplotlib.pyplot as plt
from numpy import mean
```

Para hacer las pruevas generamos de forma sintética una cadena (*w_length*) de 200 valores de longitudes de onda con valores aleatórios entre 0 y 1000:

```python
w_length = sorted([random.randint(0,1000) for i in range(0,200)])
print w_length[0:10] #Solo mostramos algunos valores a modo de ejemplo
```

Del mismo modo generamos una cadena (*values*) de valores fictícios de reflectancia con valores de 0.0 a 0.1:


```python
values = [random.random() for i in range(0,200)]
print values[0:10] #Solo mostramos algunos valores a modo de ejemplo
```

El aspecto de nuestros datos iniciales es el siguiente:

```python
plt.plot(w_length,values)
plt.ylabel('Reflectance')
plt.xlabel('Wavelength (nm)')
```

La función que declaramos a continuación nos devuelve el valor medio de los valores de reflectancia dado un valor centrar (*central_value*) y un tamaño de ventana (*window*).  
Por ejemplo, si definimos cómo valor central 100 y ventana de 100, obtendremos la media de reflectancia de las longitudes de onda entre 50 (*low*) y 150 (*high*). 

**low = central_value - (window/2) = 100 - (100/2) = 50**  
**high = central_value + (window/2) = 100 + (100/2) = 150**

```python
def mean_window(w_length,values,central_value,window):
    low=central_value-(window/2) #definimos el valor mínimo de longitud de onda para hacer la media
    high=central_value+(window/2) #definimos el valor máximo de longitud de onda para hacer la media
    
    return (mean([values[i] for i,x in enumerate(w_length) if (x <= high and x>= low)]))
```

Para entender mejor que hace la función anterior vamos algunos ejemplos de *list comprehensions*:

```python
[(i,x) for i,x in enumerate (w_length)][0:10] #mostramos los 10 primeros valores [0:10]
```

Nos devuelve una *tuple* con los valores de w_length (*x*) y su posición en el vector (*i*)

```python
[(i,x,values[i]) for i,x in enumerate (w_length)][0:10] #mostramos los 10 primeros valores [0:10]
```

Nos devuelve una *tupla* con los valores de w_length (*x*) y su posición en el vector (*i*) y además el valor de reflectancia que corresponde a cada uno valores[i].   

```python
high = 150
low = 50
[values[i] for i,x in enumerate(w_length) if (x <= high and x>= low)]
```

Nos devuelve los valores de reflectancia de las longitudes de onda que se encuentran entre 50 y 150.  

Y a continuación le aplicamos el método *mean* para obtener la media de los valores anteriores:

```python
mean([values[i] for i,x in enumerate(w_length) if (x <= high and x>= low)])
```

Hacemos un par de pruevas para comprovar que el algoritmo que hemos creado sea correcto (Los dos valores deben ser iguales).

```python
mean_window(w_length,values,15,30) 
```

```python
mean(values[0:7]) #buscamos el rango de forma manual
```

A continuación aplicamos el algoritmo generado a un número de valores (longitudes de onda) que hemos definido:

```python
new_w_length=[100,200,300,400,500,600,700,800,900]
print new_w_length
```

```python
new_w_length=range(100,1000,100) #es equivalente a la acción anterior
print new_w_length
```

**range(minimum, maximum,step)** -> Obtenemos una cadena con valores de en el rango [minimum,maximum) equidistantes un step.

```python
window=100 #definimos el valor de la ventana
```

```python
new_values=[mean_window(w_length,values,x,window) for x in new_w_length]
print new_values
```

A continuación representamos de manera gráfica los valores medios obtenidos en función del valor central de longitud de onda correspondiente en cada caso:

```python
plt.plot(new_w_length,new_values)
plt.ylabel('Reflectance')
plt.xlabel('Wavelength (nm)')
```

A continuación mostramos una gráfica con los valores obtenidos representados de una forma mucha más clara y en la segunda gráfica junto a los valores originales:

```python
plt.stem(new_w_length,new_values, '-.')
plt.plot(new_w_length,new_values)

plt.ylabel('Reflectance')
plt.xlabel('Wavelength (nm)')

plt.show()
```

```python
plt.stem(new_w_length,new_values, '-.')
plt.plot(w_length,values)
plt.plot(new_w_length,new_values)

plt.ylabel('Reflectance')
plt.xlabel('Wavelength (nm)')

plt.show()
```
